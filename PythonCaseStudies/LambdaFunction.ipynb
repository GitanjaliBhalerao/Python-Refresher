## Lambda Functions in Python â€“ Advanced Automotive Case Study
# Scenario:
An automotive company wants to analyze vehicle fuel efficiency and performance. We'll use lambda functions to filter, sort, transform, and classify data while adding some complexity, such as:

Weighted Fuel Efficiency Calculation (MPG adjusted for engine size).
Performance Index Calculation (Based on engine size, fuel efficiency, and vehicle category).
Categorization of Cars Based on Performance

# Step 1: Sample Data
We'll use a list of dictionaries, where each dictionary represents a car with attributes like model, mpg, engine_size, and horsepower.

```python
# Sample data : List of car data
cars = [
    {"model": "Toyota Corolla", "mpg": 32, "engine_size": 1.8, "horsepower": 139,"weight":1300, "fuel":"Petrol"},
    {"model": "Honda Civic", "mpg": 36, "engine_size": 1.5, "horsepower": 158,"weight":1250, "fuel":"Petrol"},
    {"model": "Ford Mustang", "mpg": 25, "engine_size": 5.0, "horsepower": 450,"weight":1500, "fuel":"Petrol"},
    {"model": "Chevrolet Camaro", "mpg": 22, "engine_size": 6.2, "horsepower": 455,"weight":1300, "fuel":"Diesel"},
    {"model": "Tesla Model 3", "mpg": 120, "engine_size": 2.0, "horsepower": 283,"weight":1800, "fuel":"Electric"},  # Electric (MPGe)
    {"model": "BMW 3 Series", "mpg": 30, "engine_size": 2.0, "horsepower": 255,"weight":1400, "fuel":"Hybrid"}
]
```

# Filtering Cars with High Fuel Efficiency
We'll use a lambda function with the filter() function to get cars with mpg greater than 30.

```python
x=filter(lambda car:car["mpg"]>30,cars)
print(x) #filter object
#print(list(x))

efficient_cars=list(filter(lambda car:car["mpg"]>30,cars))

for car in efficient_cars:
    print(car)
```

# Sorting Cars by Fuel Efficiency
We'll use sorted() with a lambda function to sort the cars in descending order of fuel efficiency.

```python
sorted_cars=sorted(cars,key= lambda car:car["mpg"])
sorted_cars
```

```python
# to sort cars in descending order of mpg
sorted_cars=sorted(cars,key= lambda car:car["mpg"],reverse=True)
for car in sorted_cars:
    print(car)
```

# Calculating Fuel Consumption in Liters per 100 km
MPG (miles per gallon) can be converted to L/100km using the formula:

L/100km = 235.214 MPG /100km


```python
for car in cars:
    car["Ltr_per_100km"]=(lambda mpg:235.214/mpg)(car["mpg"])
    print(car)

print("....................................................................................")

#for car in cars:
    #car["L_per_100km"]=round((lambda mpg:235.214/mpg)(car["mpg"]),2)
    #print(car)
```

```python
for car in cars:
    car["L_per_100km"]=round((lambda mpg:235.214/mpg)(car["mpg"]),2)
    print(car)
```

# Step 2: Advanced Filtering (Fuel-Efficient & High-Power Cars)
We filter cars that have MPG > 30 and horsepower > 150.

```python
filtered_cars=list(filter((lambda car:car["mpg"]>30 and car["horsepower"]>150),cars))

for car in filtered_cars:
    print(car)
```

# Step 3: Weighted Fuel Efficiency (Considering Engine Size)
We'll define a weighted fuel efficiency metric that considers both MPG and engine size:

WeightedÂ MPG= (MPG) /(1+EngineSize/3)


```python
for car in cars:
    car["weighted_mpg"]=round(( lambda mpg,size:mpg/(1+size/3))(car["mpg"],car["engine_size"]),2)

print("cars with  weighted fuel efficiency metric")
for car in cars:
    print(car)
```

# Step 4: Performance Index Calculation
We calculate a performance index using the following formula:

PerformanceÂ Index= (Horsepower Ã— WeightedÂ MPG) /10


```python
for car in cars:
    car["performance_index"]=round(( lambda hp,wmpg:(hp*wmpg)/10)(car["horsepower"],car["weighted_mpg"]),2)

print("Cars with Performance Index calculation")
for car in cars:
    print(car)
```

# Step 5: Categorizing Cars by Performance
We classify cars based on their performance index using a lambda function:

Low Performance: < 400
Medium Performance: 400 - 1000
High Performance: > 1000

```python
#categorize it

for car in cars:
    car["performance_category"]=(lambda index:"Low Performance" if index<400 else "Medium Performance" if index <1000 else "High Performance")(car["performance_index"])

print("Cars with performance category")

for car in cars:
    print(car["model"]," ",car["performance_index"]," ",car["performance_category"])
```
