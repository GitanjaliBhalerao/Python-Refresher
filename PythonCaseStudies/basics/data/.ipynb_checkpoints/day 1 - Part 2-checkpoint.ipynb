## Read Once Analyse Many times

```python
import csv

records=[]

# open csv file and read it row by row ==> converting to dict
with open('data/vehicle_telemetry.csv') as fl:
    reader=csv.DictReader(fl)
    for row in reader:
        records.append(row)

records[:2] # last 2 rows
```

```python
len(records)
```

## Extract features - focused 1. engine_temp and 2. engine_rpm

```python
engine_temps=[float(r["engine_temp"]) for r in records] # list comprehension

rpm_values=[int(r['engine_rpm']) for r in records]

print(engine_temps)
print(rpm_values)
```

```python
## Max engine temp
print("Maximum engine tempearature is: ",max(engine_temps))
```

```python
print("Maximum engine RPM is: ",max(rpm_values))
```

## Threashold- based event detection


```python
# constants
MAX_ENGINE_TEMP_C=105
MAX_ENGINE_RPM=6500
MIN_BATTERY_VOLTAGE=11.8
```

```python
overheat_events=[t for t in engine_temps if t> MAX_ENGINE_TEMP_C]
print("Overheat events count is ",len(overheat_events))
overheat_events
```

```python
if len(overheat_events) > 0:
    print('Vehicle needs inspection...')
else:
    print('Vehicle behaves normally.')
```

## generate summary for reporting


```python
import os
from datetime import datetime
os.makedirs("outputs",exist_ok=True)

file_name=f"outputs/telematics_summary_{datetime.now():%Y%m%d_%H%M}.txt"

with open(file_name,"w") as f:
    f.write("============================Telematics health summary =========================================\n")
    f.write(f"max_temp is {max(engine_temps)} \n")
    f.write(f"Overheat events is {len(overheat_events)} are {overheat_events}")
```

## reusable telematics valudation logic 

```python
def evaluate_vehicle_health(records):
    temps=[float(r["engine_temp"]) for r in records]
    volts=[float(r["battery_voltage"]) for r in records]

    if max(temps) > MAX_ENGINE_TEMP_C:
        return "CRITICAL : Overheating"
    if min(volts) < MIN_BATTERY_VOLTAGE:
        return "WARNING : Battery Health"

    return "NORMAL"
```

```python
status=evaluate_vehicle_health(records)
status
```

## NumPy (Numerical Python) is an open source Python library that’s used in almost every field of science and engineering. 
It’s the universal standard for working with numerical data in Python, and it’s at the core of the scientific Python and PyData ecosystems.
The NumPy API is used extensively in Pandas, SciPy, Matplotlib, scikit-learn, scikit-image and most other data science and scientific Python
packages.The NumPy library contains multidimensional array and matrix data structures
## What is NumPy?
NumPy stands for numeric python which is a python package for the computation and processing of the multidimensional and single dimensional
array elements.
Travis Oliphant created NumPy package in 2005 by injecting the features of the ancestor module Numeric into another module Numarray.
## Why use NumPy?
NumPy arrays are faster and more compact than Python lists. An array consumes less memory and is convenient to use. NumPy uses much less
memory to store data and it provides a mechanism of specifying the data types. This allows the code to be optimized even further.
