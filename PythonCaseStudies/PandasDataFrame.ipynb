```python
import numpy as np

#raw data
products=np.array(['Laptop','Smartphones','Tablets','Headphones','keyboard','Mouse','TV'])
units_sold=np.array([100,150,50,250,300,100,120])
prices=np.array([50000.00,20000.00,25000.00,2000.00,1000.00,500.00,60000.00])

#calculate revenue
revenues=units_sold*prices
print(revenues)
```

```python
#find highest revenue
max_index=np.argmax(revenues)
print(max_index)
print('Highest grossing product ',(revenues[max_index]))
#products[0]
```

```python
large_quantity=np.argmax(units_sold)
print('Large quantity : index ',large_quantity)
print('Large quantity : value ',(units_sold[large_quantity]))
```

```python

import pandas as pd
import numpy as np

# Load CSV
data = pd.read_csv("sales_transpose.csv")

# Set 'product' as index and transpose
data_t = data.set_index('product').T
print("\nCorrect Transposed Data:")
print(data_t)

# Calculate revenue
data_t['revenue'] = data_t['units_sold'].astype(int) * data_t['Price'].astype(int)
print("\nData with Revenue:")
print(data_t)
```

```python
max_row=data_t.loc[data_t['revenue'].idxmax()]
print(data)
max_row
```
